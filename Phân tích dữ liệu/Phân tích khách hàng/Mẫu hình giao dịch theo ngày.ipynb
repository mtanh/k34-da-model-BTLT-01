```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../datasets/Invoices_Std.csv", encoding='latin-1', low_memory=False)
```

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

tx_daily = df.groupby(df["InvoiceDate"].dt.date)["InvoiceNo"].nunique()

plt.figure(figsize=(10,4))
plt.plot(tx_daily.index, tx_daily.values, marker="o")
plt.xlabel("Ngy")
plt.ylabel("S n hng")
plt.title("Tn sut giao dch hng ngy")
plt.grid(True)
plt.tight_layout()
plt.show()
```
