```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../datasets/Invoices_Std.csv", encoding='latin-1', low_memory=False)
```

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df["YearMonth"] = df["InvoiceDate"].dt.to_period("M")

tx_monthly = df.groupby("YearMonth")["InvoiceNo"].nunique()

plt.figure(figsize=(10,4))
plt.plot(tx_monthly.index.to_timestamp(), tx_monthly.values, marker="o")
plt.xlabel("Tháng")
plt.ylabel("Số đơn hàng")
plt.title("Tần suất giao dịch hàng tháng")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Tháng 1 đến 2 sức mua giảm.

Tháng 3 đến tháng 6 sức mua tăng nhẹ rồi ổn định.

Từ tháng 9 đến tháng 11 sức mua tăng mạnh nhất cả năm.
