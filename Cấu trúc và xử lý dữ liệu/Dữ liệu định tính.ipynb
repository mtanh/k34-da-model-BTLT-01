```python
import pandas as pd
from tabulate import tabulate
```

```python
Invoices = pd.read_csv("../datasets/Invoices.csv", encoding='latin-1')

columns = {
    "Invoices": ["Description", "Country"]
}

def profile_categorical(df, file_name, col_name):
    total = len(df)
    missing = df[col_name].isna().sum()
    distinct = df[col_name].nunique(dropna=True)
    cardinality = round(distinct / total, 3) if total > 0 else 0

    freq = df[col_name].value_counts(dropna=True)
    mode = freq.idxmax()
    mode_pct = round(freq.max() / total * 100, 2)

    return {
        "Tập tin": file_name,
        "Cột": col_name,
        "Tổng": total,
        "Thiếu (%)": round(missing / total * 100, 2),
        "Giá trị riêng biệt": distinct,
        "Độ phân biệt (Cardinality)": cardinality,
        "Giá trị phổ biến (Mode)": mode,
        "Tỷ lệ Mode (%)": mode_pct
    }

results = []
for file, cols in columns.items():
    for col in cols:
        df = eval(file)
        results.append(profile_categorical(df, file, col))

categorical_profile = pd.DataFrame(results)

table_text = tabulate(categorical_profile, headers='keys', tablefmt='grid', showindex=False)
print(table_text)
```
